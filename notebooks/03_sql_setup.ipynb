# H1B Visa Sql Setup

## Step-1 Import libraries

```python
import pandas as pd
from sqlalchemy import create_engine
```

## Step 2 — Create MySQL Database Connection
### Why?
We connect Python to the local MySQL server using username and password.

```python
engine = create_engine("mysql+pymysql://root:mahin%40123@localhost:3306/h1b_project")
```

## Step 3 — Load Clean CSV File
### Why?
We read the cleaned dataset created in the previous notebook.

```python
df = pd.read_csv("../data/processed/h1b_clean_2024.csv")
print(df.shape)
```

## Step 4 — Upload Data to MySQL Table
### Why?
We store the cleaned dataset inside MySQL for SQL analysis.

```python
df.to_sql(
    name="h1b_clean",
    con=engine,
    if_exists="replace",
    index=False,
    chunksize=5000
)

print("Data uploaded successfully!")
```

## Step 5 — Verify Upload
### Why?
We check if the table exists and confirm row count.

```python
query = "SELECT COUNT(*) AS total_rows FROM h1b_clean;"
pd.read_sql(query, engine)
```
