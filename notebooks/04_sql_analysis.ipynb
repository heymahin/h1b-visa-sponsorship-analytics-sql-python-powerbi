# H1B Visa Sql Analysis

```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine("mysql+pymysql://root:mahin%40123@localhost:3306/h1b_project")
```

## Analysis 1 — Top Sponsoring Employers
### Why?
Companies that file the most applications are strong targets for H1B job seekers.

```python
query1 = """
SELECT EMPLOYER_NAME, COUNT(*) AS total_filings
FROM h1b_clean
GROUP BY EMPLOYER_NAME
ORDER BY total_filings DESC
LIMIT 10;
"""

top_employers = pd.read_sql(query1, engine)
top_employers
```

## Analysis 2 — Highest Paying Job Roles
### Why?
Understanding which job roles offer the highest average salaries helps students choose career paths with stronger earning potential and better sponsorship opportunities.

### Business Value:
Applicants can prioritize high-value roles such as Data Science, Cloud Engineering, AI, and Cybersecurity.

```python
query2 = """
SELECT JOB_TITLE, ROUND(AVG(ANNUAL_SALARY),2) AS avg_salary
FROM h1b_clean
GROUP BY JOB_TITLE
HAVING COUNT(*) >= 20
ORDER BY avg_salary DESC
LIMIT 10;
"""

pd.read_sql(query2, engine)
```

## Analysis 3 — Best States for H1B Opportunities
### Why?
Some states have significantly more H1B job opportunities due to the concentration of technology companies, consulting firms, and large enterprises.

### Business Value:
Applicants can target states with higher demand to improve their chances of finding sponsorship opportunities.

```python
query3 = """
SELECT WORKSITE_STATE, COUNT(*) AS total_jobs
FROM h1b_clean
GROUP BY WORKSITE_STATE
ORDER BY total_jobs DESC
LIMIT 10;
"""

pd.read_sql(query3, engine)
```

## Analysis 4 — Job Concentration by State
### Why?
Different states specialize in different job markets. Identifying which roles are most common in each state helps applicants align their job search geographically.

### Business Value:
Students can apply strategically by matching their skills to the strongest job markets in each state.

```python
query4 = """
SELECT WORKSITE_STATE, JOB_TITLE, COUNT(*) AS total
FROM h1b_clean
GROUP BY WORKSITE_STATE, JOB_TITLE
ORDER BY total DESC
LIMIT 20;
"""

pd.read_sql(query4, engine)
```

## Analysis 5 — Top Industries Hiring H1B Workers
### Why?
Identifying industries with the highest number of filings helps applicants focus on sectors that actively sponsor international talent.

```python
query5 = """
SELECT NAICS_CODE, COUNT(*) AS total_filings
FROM h1b_clean
GROUP BY NAICS_CODE
ORDER BY total_filings DESC
LIMIT 10;
"""

pd.read_sql(query5, engine)
```

## Analysis 6 — Average Salary by State
### Why?
Comparing salaries across states helps applicants balance opportunity with compensation.

```python
query6 = """
SELECT WORKSITE_STATE,
ROUND(AVG(ANNUAL_SALARY),2) AS avg_salary
FROM h1b_clean
GROUP BY WORKSITE_STATE
ORDER BY avg_salary DESC
LIMIT 10;
"""

pd.read_sql(query6, engine)
```

## Analysis 7 — Visa Class Distribution
### Why?
This shows which visa categories dominate the dataset and confirms focus on H1B-related filings.

```python
query7 = """
SELECT VISA_CLASS, COUNT(*) AS total
FROM h1b_clean
GROUP BY VISA_CLASS
ORDER BY total DESC;
"""

pd.read_sql(query7, engine)
```

## Analysis 8 — Full-Time Position Analysis
### Why?
Most H1B roles are full-time. This confirms the type of employment commonly offered.

```python
query8 = """
SELECT FULL_TIME_POSITION, COUNT(*) AS total
FROM h1b_clean
GROUP BY FULL_TIME_POSITION;
"""

pd.read_sql(query8, engine)
```

## Analysis 9 — Quarterly Filing Trend
### Why?
Quarterly trends help identify seasonal patterns in H1B filings during FY2024.

```python
query9 = """
SELECT quarter, COUNT(*) AS total_filings
FROM h1b_clean
GROUP BY quarter
ORDER BY quarter;
"""

pd.read_sql(query9, engine)
```

## Analysis 10 — Highest Paying Employers
### Why?
Applicants can target companies that consistently offer higher salaries.

```python
query10 = """
SELECT EMPLOYER_NAME,
ROUND(AVG(ANNUAL_SALARY),2) AS avg_salary,
COUNT(*) AS total_jobs
FROM h1b_clean
GROUP BY EMPLOYER_NAME
HAVING COUNT(*) >= 20
ORDER BY avg_salary DESC
LIMIT 10;
"""

pd.read_sql(query10, engine)
```
